```python
# calculate reliability between two runs:
  
def FormatSubID(source_id):
    # Make it a string
    source_id = str(source_id)

    if "bio" in source_id:   
        try:
            # check if the sub number is four digit:
            sub_num = int(source_id.split("bio")[1])
        except ValueError:
            sub_num = "0000"
        
    elif "Bio" in source_id:
        try:
            # check if the sub number is four digit:
            sub_num = int(source_id.split("Bio")[1])
        except ValueError:
            sub_num = "0000"    
        
    elif "BIO-" in source_id:
        try:
            # check if the sub number is four digit:
            sub_num = int(source_id.split("BIO-")[1])
        except ValueError:
            sub_num = "0000"

    elif "BIO" in source_id:
        try:
            sub_num = int(source_id.split("BIO")[1])
        except ValueError:
            sub_num = "0000"   
    
    else: # everything else just empty
        sub_num = "0000"
    if sub_num != "0000":
        formatted_id = "bio" + "{:04}".format(sub_num)
    elif sub_num == "0000":
        formatted_id = ""
    else:
        formatted_id = ""
    return formatted_id
```
